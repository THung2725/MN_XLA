### 1. VIẾT CHƯƠNG TRÌNH BIẾN ĐỔI ẢNH


##### 1.1. Chọn đối tượng trong ảnh

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("fruit.jpg")
bmg = data[800:1200, 570:980]
print(data.shape)

iio.imsave("orange.jpg", bmg)
plt.imshow(bmg)
plt.show()
```

##### 1.2 Tịnh tiến đơn


```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("fruit.jpg", mode = "F")
bdata = nd.shift(data,(100,25))

plt.imshow(bdata)
plt.show()
```

##### 1.3. Thay đổi kích thước ảnh

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("fruit.jpg")
print(data.shape)
bdata = nd.zoom(data,2)
print(bdata.shape)
bdata2 = nd.zoom(data, (2, 2, 1))
print(bdata2.shape)
data3 = nd.zoom(data, (0.5, 0.9, 1))
plt.imshow(data3)
plt.show()
```

##### 1.4. Xoay ảnh

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("fruit.jpg", mode = "F")
print(data.shape)
d1 = nd.rotate(data, 20)
plt.imshow(d1)
plt.show()

d2 = nd.rotate(data, 20, reshape=False)
plt.imshow(d2)
plt.show()
```

##### 1.5. Dilation và Erosion

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("world_cup.jpg", mode = "L")
print(data.shape)

d1 = nd.binary_dilation(data)
plt.imshow(d1)
plt.show()

d2 = nd.binary_dilation(data, iterations=3)
plt.imshow(d2)
plt.show()
```

##### 1.6. Coordinate Mapping

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("world_cup.jpg", mode = "F")
print(data.shape)
V, H = data.shape

M = np.indices((V, H))

d = 5 
q = 2 * d * np.random.rand(*M.shape) - d

mp = (M + q).astype(int)
d1 = nd.map_coordinates(data, mp)
plt.imshow(d1)
plt.show()
```

##### 1.7. Biến đổi chung(Generic Tranformation)



```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

def GeoFun(outcoord):
    a = 10 * np.cos(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.cos(outcoord[1] / 10.0) + outcoord[1]
    return a, b

data = iio.imread("world_cup.jpg", mode = "F")

d1 = nd.geometric_transform(data, GeoFun)

plt.imshow(d1)
plt.show()
```

### 2. BÀI TẬP

##### 1. Viết chương trình chọn quả kiwi từ ảnh colorfull-ripe-tropical-fruit.jpg trong thư mục exercise. Tịnh tiến quả kiwi sang 30 pixel

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("exercise/colorful-ripe-tropical-fruits.jpg")
bmg = data[920:1100, 390:570]
bdata = nd.shift(bmg, (0, 30, 0))

print(data.shape)
iio.imsave("kiwi.jpg", bdata)

plt.imshow(bdata)
plt.show()
```

##### 2. Viết chương trình chọn quả đu đủ và quả dưa hấu từ ảnh colorfull-ripe-tropical-fruit.jpg trong thư mục exercise. đổi màu hai đối tượng này

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

data = iio.imread("exercise/colorful-ripe-tropical-fruits.jpg")
bmg = data[350:800, 120:670]
bmg2 = data[270:1100, 1630:2100]

bmg[:, :, 1] = 0
bmg2[:, :, 0] = 0

print(data.shape)
iio.imsave("dudu.jpg", bmg)
iio.imsave("dua_hau.jpg", bmg2)

plt.imshow(bmg)
plt.show()
plt.imshow(bmg2)
plt.show()
```
